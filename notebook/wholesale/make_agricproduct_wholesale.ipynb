# 데이터 불러오기

```python
import pandas as pd
```

```python
master_df = pd.read_csv('./result/wholesale_2022-06-23_preprocessed.csv')
```

```python
master_df
```

```python
s = "흰 콩(수입)(35kg)"
s.__contains__("수입")
```

```python
def is_domestic_or_imported(s: str) -> bool:
    """문자열에 수입이 포함되면 False 아니면 True"""
    return "수입" if s.__contains__("수입") else "국산"
```

```python
master_df['kind'] = master_df['kind_name_new'].apply(is_domestic_or_imported)
```

```python
# 국산 수입 비율
master_df['kind'].value_counts()
```

```python
master_df
```

```python
agricproduct_wholesale_df = master_df[['category_code', 'kind', 'p_group', 'rank', 'unit']]
```

```python
agricproduct_wholesale_df
```

```python
agricproduct_wholesale_df = agricproduct_wholesale_df.drop_duplicates().reset_index(drop=True)
```

```python
agricproduct_wholesale_df.to_csv('./result/agriproduct_wholesale_df.csv', index=False)
```
