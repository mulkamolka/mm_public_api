# 상품 코드 생성

- 상품을 식별할 수 있는 테이블의 기본키를 생성해야한다.
- 상품명, 상품 description, 카테고리코드, 상점코드가 같은 경우 같은 상품으로 식별

````python
p_code_lst = []
prev_dict = dict()
cnt = 1

for idx, row in master_df.iterrows():
    
    key = (row['p_name'], row['description'], row['category_code'], row['m_code'])

    if key in prev_dict:
        # 이미 등록된 상품코드
        p_code = prev_dict[key]
        
    else:
        # 새로운 상품코드 생성
        p_code = "AR" + str(cnt).zfill(6)
        prev_dict[key] = p_code
        cnt += 1
        
    p_code_lst.append(p_code)
````

````python
len(p_code_lst)
````

````python
master_df['p_code'] = p_code_lst
````

````python
len(master_df['p_code'].unique())
````
