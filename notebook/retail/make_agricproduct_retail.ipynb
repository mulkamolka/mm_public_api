# market_retail에 없는 상점을 df에서 제거

```python
market_df = pd.read_csv('./result/market_retail.csv', encoding='utf8')
```

```python
market_df
```

```python
# 폐업한 상점을 리스트에서 제거 
df = df[df['M_NAME'].isin(market_df['m_name'])].reset_index(drop=True)
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
# 농산물 리스트
agri_lst = ['사과', '배', '배추', '무', '양파', '상추', '오이', '애호박', '호박', "쌀"]
```

```python
# A_NAME이 결측값인 레코드를 삭제하고 문자열처리를 통해 p_group을 생성한다
df = df.dropna(subset=['A_NAME'])
df['p_group'] = df['A_NAME'].str.split('(')
df['p_group'] = df['p_group'].apply(lambda x : x[0])
```

```python
# 소매점포에서 판매하는 농산물 상품 데이터 
agri_df = df[df['p_group'].isin(agri_lst)]
```

```python
agri_df.head()
```

# 상점코드 매핑

```python
market_df
```

```python
market_dict = dict(zip(market_df.m_name, market_df.m_code))
```

```python
market_dict
```

```python
def mapping_m_code(s):
    return market_dict.get(s)
```

```python
agri_df['m_code'] = agri_df['M_NAME'].apply(mapping_m_code)
```

```python
agri_df.head()
```

# description 생성

- ADD_COL과 A_NAME을 결합하여 결측치 제거

```python
agri_df.isnull().sum()
```

```python
# 상품명에서 ( 다음 문자 합치기 및 마지막 괄호 제거
agri_df['description'] = agri_df['A_NAME'].str.split('(')
agri_df['description'] = agri_df['description'].apply(lambda x: " ".join(x[1:]).replace(')', ''))

# 비고가 공백인 경우 blank 처리
agri_df['ADD_COL'] = agri_df['ADD_COL'].fillna("")

agri_df['description'] = agri_df['ADD_COL'] + ' ' + agri_df['description']
```

```python
agri_df
```

```python
agri_df = agri_df.reset_index(drop=True)
```

```python
agri_df.isnull().sum()
```

```python
agri_df['p_group'].unique()
```

```python
# 농산물 그룹식별을 위한 아스키코드값을 만들어 값으로 설정한다. e.g 사과 : A
agri_dict = {item : chr(idx + 65) for idx, item in enumerate(agri_lst)}
```

```python
agri_dict
```

### 농산품 그룹 식별 코드를 CSV 파일로 저장

```python
# 해당 알파벳 코드에 해당하는 그룹의 모든 단위를 저장한다.

uniq_units = {}

for key in agri_dict.keys():
    uniq_units[agri_dict[key]] = list(agri_df.query(f"p_group == '{key}'")['A_UNIT'].unique())
```

```python
# 사과의 모든 단위를 출력
# print(uniq_units['A'])
```

```python
for key, val in agri_dict.items():
    df = pd.DataFrame(data=uniq_units[val], columns=[key])
    df.to_csv(f"./units/agr-{val}.csv", index=False)
```

```python
df.head()
```

# 농작물 상품카테고리코드 생성

- e.g) AGR-A0001 : 사과의 첫번째 단위

- p_group과 단위를 기준으로 unique한 카테고리 코드를 생성

```python
agri_df.head()
```

```python
agriproduct_retail = {
    'category_code' : [],
    'p_group' : [],
    'unit' : [],
    'price': [],
    'm_code': [],
    'p_name': [],
    'date' : [],
    'description': []
}
```

```python
for key in agri_dict.keys(): # 사과, 배, 배추 ...
    
    q = f"p_group == '{key}'"
    print(f"------------------{key}------------------------")
    
    ###########################################################
    #        농산물 소매 테이블 정보 저장
    ###########################################################
    for idx, row in agri_df.query(q).iterrows():

        agriproduct_retail['p_name'].append(row[4])
        agriproduct_retail['unit'].append(row[5])
        agriproduct_retail['price'].append(row[6])
        agriproduct_retail['description'].append(row[8])
        agriproduct_retail['date'].append(row[9])
        agriproduct_retail['p_group'].append(row[14])
        agriproduct_retail['m_code'].append(row[15])

        ##########################################################
        #       Category_code 생성
        ##########################################################
        
        # 단위 딕셔너리를 생성한다.
        unit_dict = { v : k + 1 for k, v in enumerate(uniq_units[agri_dict[key]])}

        # unit으로  unit_dict 딕셔너리에서 해당 unitcode를 검색한다.
        unitcode = str(unit_dict[row[5]])

        # agr-A0001 형식으로 category_code를 생성한다.
        # agr : 농산물 소매테이블
        # A : 사과
        # 0001 : 1번째 unitcode
        category_code = "AGR-" + agri_dict[key] + unitcode.zfill(4)

        # agriproduct_retail_df에 category_code 값을 저장한다.
        agriproduct_retail_df['category_code'].append(category_code)
        
print("--------------------종료-----------------------")

agriproduct_retail_master_df = pd.DataFrame(agriproduct_retail)
```

```python
# 농산물 소매상품 마스터 테이블
agriproduct_retail_master_df
```

```python
agriproduct_retail_df = agriproduct_retail_master_df[['category_code', 'p_group', 'unit']]
```
